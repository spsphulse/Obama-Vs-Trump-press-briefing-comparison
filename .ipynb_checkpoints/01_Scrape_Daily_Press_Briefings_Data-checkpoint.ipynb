```python
#Bread & butter
import pandas as pd
from itertools import product
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
from selenium.webdriver.chrome.options import Options


```

```python
# define URLs & datapoints to scrape
obama_briefing_url='https://2009-2017.state.gov/r/pa/prs/dpb/'
trump_briefing_url='https://www.state.gov/r/pa/prs/dpb/'
obamaYearList=['2013','2014','2015','2016']
trumpYearList=['2017','2018']
monthList=['01','02','03','04','05','06','07','08','09','10','11','12']
briefing_df=pd.DataFrame(columns=['Year','Month','Day','President','Tick_Color'])


# Set basic driver options. Minimum logging
chrome_options = Options()
chrome_options.add_argument('--log-level=3')
driver = webdriver.Chrome("C:\chromedriver_win32\chromedriver.exe")
```

```python
# Scrape data for 'Daily' press briefings for first 2 years of Obama administration

obamaYearMonth=list(product(obamaYearList,monthList))

for yearMonth in obamaYearMonth:
    print(obama_briefing_url+yearMonth[0]+"/"+yearMonth[1])
    link=obama_briefing_url+yearMonth[0]+"/"+yearMonth[1]
    driver.get(link)
    allDays=driver.find_elements_by_xpath('//li[@class="day"]')
    allBriefingDays=driver.find_elements_by_xpath('//a[@class="view"]')
    print(str(len(allDays))+"----------"+str(len(allBriefingDays)))
    dayNum=0
    for day in allDays:
        dayNum+=1
        #print(day.find_elements_by_class_name('view'))
        if((day.find_elements_by_class_name('view'))):            
            print(str(dayNum)+'------ Press Briefing Present')
            briefing_df.loc[len(briefing_df)]=[yearMonth[0],yearMonth[1],dayNum,'Obama','Blue']
    
    
  
    
```

```python
# Scrape data for 'Daily' press briefings for first 2 years of Obama administration

trumpYearMonth=list(product(trumpYearList,monthList))

for yearMonth in trumpYearMonth:
    print(trump_briefing_url+yearMonth[0]+"/"+yearMonth[1])
    link=trump_briefing_url+yearMonth[0]+"/"+yearMonth[1]
    driver.get(link)
    allDays=driver.find_elements_by_xpath('//li[@class="day"]')
    allBriefingDays=driver.find_elements_by_xpath('//a[@class="read"]')
    print(str(len(allDays))+"----------"+str(len(allBriefingDays)))
    dayNum=0
    for day in allDays:
        dayNum+=1
        #print(day.find_elements_by_class_name('view'))
        if((day.find_elements_by_class_name('read'))):            
            print(str(dayNum)+'------ Press Briefing Present')
            briefing_df.loc[len(briefing_df)]=[yearMonth[0],yearMonth[1],dayNum,'Trump','Red']
    
```

```python
#Save data for plotting

briefing_df.to_csv('Full_Press_Briefing_Data.csv',index=False)
```
